# Introduccion a Matplotlib
### Libreria utilizada en python para el analisis de datos pero de forma grafica
### como pueden ser graficos de linea, de dispersion,de barras y sectores, 

```python
import matplotlib.pyplot as plt
```

# Grafico de Linea

```python
#  Grafco de Linea
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='o')
plt.title("Gráfico de Líneas")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

# Gráfico de dispersión

```python
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color='green')
plt.title("Gráfico de Dispersión")
plt.xlabel("Edad")
plt.ylabel("Puntaje")
plt.show()
```

# Gráfico de barras

```python
categorias = ['A', 'B', 'C', 'D']
valores = [23, 17, 35, 29]

plt.bar(categorias, valores, color='skyblue')
plt.title("Gráfico de Barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

 # Gráfico de pastel

```python
etiquetas = ['Python', 'Java', 'C++', 'Ruby']
porcentajes = [40, 25, 20, 15]

plt.pie(porcentajes, labels=etiquetas, autopct='%1.1f%%', startangle=140)
plt.title("Lenguajes de Programación Preferidos")
plt.axis('equal')  #  círculo perfecto
plt.show()
```

# Guardando un grafico como imagen con la libreria matplotlib


```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Gráfico Guardado")
plt.savefig("grafico.png")  #  gráfico como archivo PNG
plt.show()
```
